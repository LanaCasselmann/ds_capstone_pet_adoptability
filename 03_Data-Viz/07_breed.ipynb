#### Combine first two into fast adoption and the last two into slow adoption

```python
df_eda_2 = breed_df.copy()
```

```python
df_eda_2.adoptionspeed.value_counts()
```

```python
# combining first two adoption speed 
df_eda_2['adoptionspeed'] = df_eda_2['adoptionspeed'].replace(1, 0)
df_eda_2['adoptionspeed'] = df_eda_2['adoptionspeed'].replace(2, 0)
```

```python
# combining last two adoption speed 
df_eda_2['adoptionspeed'] = df_eda_2['adoptionspeed'].replace(3, 1)
df_eda_2['adoptionspeed'] = df_eda_2['adoptionspeed'].replace(4, 1)
```

```python
pd.crosstab(df_eda_2.breed_type,df_eda_2.adoptionspeed,normalize="index").plot.bar(xlabel='Breed', ylabel='Proportion')
x_labels = ['Pure', 'Mix']
plt.ylim(0.0, 0.9)
plt.xticks(ticks=np.linspace(0,1,2),labels=x_labels, rotation = 0)
ax = plt.gca()
legend = ax.legend(title="Adoption speed", loc="upper center",frameon=False, labelcolor='#365b6d')
labellist2 = ["First week/month", "First 3 months/No adoption"]
for i,label in enumerate(legend.get_texts()):
    #label.set_weight('bold')
    label.set_text(labellist2[i]) 

plt.savefig('../images/pure_hybrid_breed.jpg', dpi=1200, bbox_inches='tight')
```

```python
pd.crosstab(df_eda_2.age_bin,df_eda_2.adoptionspeed,normalize="index").plot.bar(xlabel='Age', ylabel='Proportion')
x_labels = ['Newborn', 'Puppy/Kitten', 'Adult', 'Senior']
plt.ylim(0.0, 0.9)
plt.xticks(ticks=np.linspace(0,3,4),labels=x_labels,color='#365b6d', rotation = 0)
ax = plt.gca()
legend = ax.legend(title="Adoption speed", loc='best', frameon=False, labelcolor='#365b6d')
labellist2 = ["First week/month", "First 3 months/No adoption"]
for i,label in enumerate(legend.get_texts()):
    #label.set_weight('bold')
    label.set_text(labellist2[i]) 

plt.savefig('../images/age_adoption_speed.jpg', dpi=1200, bbox_inches='tight')
```
